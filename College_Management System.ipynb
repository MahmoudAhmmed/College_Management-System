# ***Hostel***

```python
from abc import ABC, abstractmethod
class Hostel(ABC):
    def __init__(self,  student_id, Block_number, Room_number):
        self.student_id = student_id
        self.Block_number = Block_number
        self.Room_number = Room_number

    @abstractmethod
    def HostelDetails(self):
        pass

    @abstractmethod
    def CheckIn(self):
        pass

    @abstractmethod
    def CheckOut(self):
        pass
        

class BoysHostel(Hostel):
    def __init__(self, student_id, Block_number, Room_number, Gender):
        super().__init__(student_id, Block_number, Room_number)
        self.Gender = Gender

    def HostelDetails(self):
        print(f"student is {self.Gender}, has ID {self.student_id}, lives in block number {self.Block_number} and room number {self.Room_number}")

    def CheckIn(self):
        print(f"Student who has ID {self.student_id} checked into room {self.Room_number}.")

    def CheckOut(self):
        print(f"Student who has ID {self.student_id} checked out of room {self.Room_number}.")
    

class GirlsHostel(Hostel):
    def __init__(self, student_id, Block_number, Room_number, Gender):
        super().__init__( student_id, Block_number, Room_number)
        self.Gender = Gender

    def HostelDetails(self):
        print(f"student is {self.Gender}, has ID {self.student_id}, lives in block number {self.Block_number} and room number {self.Room_number}")

    def CheckIn(self):
        print(f"Student who has ID {self.student_id} checked into room {self.Room_number}.")

    def CheckOut(self):
        print(f"Student who has ID {self.student_id} checked out of room {self.Room_number}.")



Ahmed = GirlsHostel("2022651", 5, 2, "male")
Ahmed.HostelDetails()
Ahmed.CheckIn()
Ahmed.CheckOut()

print("-"*100)
print("-"*100)

Menna = GirlsHostel("2022651", 5, 2, "Female")
Menna.HostelDetails()
Menna.CheckIn()
Menna.CheckOut()
print("-"*100)

college.add_Hostels(Ahmed) # Agreggation 
```

# ***Student***

```python
from abc import ABC, abstractmethod
class Student(ABC):
    def __init__(self, student_id, student_name, gender, Age, class_id):
        self.student_id = student_id
        self.student_name = student_name
        self.gender = gender
        self.Age = Age
        self.class_id = class_id
        self.attendence = []

    @abstractmethod
    def student_details(self):
        pass

    @abstractmethod
    def payfees(self):
        pass


class UGStudent(Student):
    def __init__(self, student_id, student_name, gender, Age, class_id):
        super().__init__(student_id, student_name, gender, Age, class_id)

    def student_details(self):
        print(f"the '{self.gender}' student '{self.student_name}', his age is '{self.Age}' years old with student ID '{self.student_id}' with Class ID '{self.class_id}'")

    def payfees(self):
        while True:
            fees = int(input("plese enter 3000 pound fees of study: "))
            if fees == 3000:
                print(f"The fees for the student {self.student_name} have been paid successfully.")
                break
                
            elif fees < 3000:
                print(f"the money is not enough please enter 3000 pound")
                
    def Attendence(self,  student_name, student_id):
        self.attendence.append([student_id, student_name])
        return "the attendence took successfully"
    
    def IsPresent(self, student_name, student_id):
        for lst_of_student in self.attendence:
            if student_id in lst_of_student:
                return f"{student_name} attend today"
        return f"{student_name} didn't attend today"


class PGStudent(Student):
    def __init__(self, student_id, student_name, gender, Age, class_id, Graduation_Year):
        super().__init__(student_id, student_name, gender, Age, class_id)
        self.Graduation_Year = Graduation_Year

    def student_details(self):
        message =  f""" 
Name: {self.student_name}
ID: {self.student_id}
Class ID: {self.class_id}
Gender: {self.gender}
Age: {self.Age}'   
Graduation Year: {self.Graduation_Year}"""

        print(message)

    def payfees(self):
        while True:
            fees = int(input("You should to pay fees to can graduate, the amount of fees is 3000: "))

            if fees == 3000:
                print(f"The fees for the post student '{self.student_name}' have been paid successfully. Congratulations on your graduation.")
                break 
            elif fees < 3000:
                print(f"the money is not enough please enter 3000 pound to can graduate")


under_graduated = UGStudent("202401", "Amir", "Male", 21, "5011")
under_graduated.student_details()

print(under_graduated.Attendence("Amir", "202401"))
print(under_graduated.IsPresent("Amir", "202401"))
under_graduated.payfees()
print("-"*100)
print("-"*100)


graduated = PGStudent("202981", "Aya", "Female", 20, "5022", "2022")
graduated.student_details()
graduated.payfees()
```

# ***Staff***

```python
from abc import ABC, abstractmethod
class Staff(ABC):
    def __init__(self, staff_id, staff_name, salary, department_id):
        self.staff_id = staff_id
        self.staff_name = staff_name
        self.salary =  salary
        self.department_id = department_id

    @abstractmethod
    def StaffDetails(self):
        pass


class TeachingStaff(Staff):
    def __init__(self, staff_id, staff_name, salary, department_id):
        super().__init__(staff_id, staff_name, salary, department_id)
    
    def StaffDetails(self):
        print(f"Staff name: {self.staff_name}  Staff id: {self.staff_id}  Department ID: {self.department_id}  Salary: {self.salary}")
        

class NonTeachingStaff(Staff):
    def __init__(self, staff_id, staff_name, salary, department_id, job):
        super().__init__(staff_id, staff_name, salary, department_id)
        self.job = job

    def StaffDetails(self):
        print(f"staff name: {self.staff_name}  staff id: {self.staff_id}  Job: {self.job}  Department ID: {self.department_id}  salary: {self.salary}")
    


teacher = TeachingStaff("502", "Zyad", 15000, "Ma001")
teacher.StaffDetails()
print("-"*100)

manger = NonTeachingStaff("501", "Motaz", 20000, "Ma011", "Head of the Math Department")
manger.StaffDetails()
```

# ***Library***

```python
class Library:
    def __init__(self, Library_id, librarian_name, Book_Selection, Total_Books):
        self.Library_id = Library_id
        self.librarian_name = librarian_name
        self.Book_Selection = Book_Selection
        self.Total_Books = Total_Books
        self.lent_books = {}


    def LibraryDetails(self):
        message = f"Library ID: {self.Library_id}\nLibrarian Name: {self.librarian_name}\nTotal Books: {self.Total_Books}\nBook Selction: {[book for book in self.Book_Selection]}\nTotal_Books: {self.Total_Books}"
        return message 


    def SearchBooks(self, book_name):
        found_book = [book for book in self.Book_Selection if book_name.lower() in book.lower()]
        if found_book:
            return f"The Book {found_book} is found"
        else:
            return "Noo Book founds with that name"


    def LendBooks(self, book_name, borrower_name):
        if book_name in self.Book_Selection  and  book_name not in self.lent_books:
            self.lent_books[book_name] = borrower_name
            return f"Book '{book_name}' lent to {borrower_name}."

        elif book_name in self.lent_books:
            return f"Book: '{book_name}' are lent to '{self.lent_books[book_name]}' "

        else:
            return f"we don't have the Book: {book_name} in our books selection"


    def ReturnBooks(self, book_name):
        if book_name in self.lent_books:
            borrower_name = self.lent_books.pop(book_name)
            print(f"The book {book_name} is returned by {borrower_name}")

        else:
            print(f"the Book {book_name} It was not on the list of lent books.")


    def PayFine(self, borrower_name, amount_money):
        print(f"{borrower_name} paid a fine of ${amount_money}.")



library = Library(
    Library_id=101,
    librarian_name="Jane Doe",
    Book_Selection=["Harry Potter", "The Hobbit", "1984", "Pride and Prejudice"],
    Total_Books=100
)

print(library.LibraryDetails())
print("-"*100)

print(library.SearchBooks("Harry"))
print("-"*100)


print(library.LendBooks("Harry Potter", "Ammar"))
print(library.LendBooks("Harry Potter", "Ahmed"))
print("-"*100)

library.ReturnBooks("Harry Potter")
print("-"*100)


library.PayFine("Ali", 5)
```

# ***Bus***

```python
class Bus:
    def __init__(self, Bus_id, Bus_number, Driver_Name, Destination, Total_Seats):
        self.Bus_id = Bus_id
        self.Bus_number = Bus_number
        self.Driver_Name = Driver_Name
        self.Destination = Destination
        self.Total_Seats = Total_Seats
    

    def BusDeatails(self):
        print(f"Bus ID: {self.Bus_id}\nBus Number :  {self.Bus_number}\nDriver_Name: {self.Driver_Name}\nDestination: {self.Destination}\nTotal Seats: {self.Total_Seats}")

    def SeatsAvailability(self):
        num_seets_book = int(input("How manny number of seats do you wnat?: "))
        if num_seets_book > self.Total_Seats:
            print("The number of seats required is greater than the bus capacity.")

        elif num_seets_book == self.Total_Seats:
            print(f"{num_seets_book} seats were reserved on the bus. the bus now is complete")

        else:
            print(f"{num_seets_book} seats were reserved on the bus. There are still {self.Total_Seats - num_seets_book} empty seats in the bus.")


Bus1 = Bus("01011", 2, "Osama", "to down town", 20)
Bus1.BusDeatails()
print("-"*100)
Bus1.SeatsAvailability()
```
